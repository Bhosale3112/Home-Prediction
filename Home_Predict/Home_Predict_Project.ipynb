Fill missing Data using Mean and Mode function

```python
cat=[]
con=[]

for i in A.columns:
    if (A[i].dtype == 'object'):
        cat.append(i)
        x=A[i].mode()[0]
        A[i]=A[i].fillna(x)
    else:
        con.append(i)
        x=A[i].mean()
        A[i]=A[i].fillna(x)
```

# Define X & Y

```python
X =A.drop(labels=["Id","SalePrice"],axis=1)
Y =A[["SalePrice"]]
```

```python
cat2=[]
con2=[]

for i in X.columns:
    if (X[i].dtype == 'object'):
        cat2.append(i)
    else:
        con2.append(i)
```

# Standardize Data:
Using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
ss =StandardScaler()
X1 =pd.DataFrame(ss.fit_transform(X[con2]),columns=con2)
```

# Find Outliers and Remove it

```python
ol=[]
for i in con2:
    ol.extend(list(X1[(X1[i]>3) | (X1[i]<-3)].index))
```

```python
from numpy import unique
ol=unique(ol)
```

```python
X=X.drop(index=ol,axis=0)
Y=Y.drop(index=ol,axis=0)
```

```python
X.index=range(0,1021,1)
```

```python
Y.index=range(0,1021,1)
```

# Preprocessing, Split the Data into training and testing

```python
from sklearn.preprocessing import StandardScaler
ss =StandardScaler()
X1 =pd.DataFrame(ss.fit_transform(X[con2]),columns=con2)
X2 =pd.get_dummies(X[cat2])
Xnew = X1.join(X2)

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=31)
```

# Use OLS model:

```python
from statsmodels.api import OLS,add_constant
xcon=add_constant(xtrain)
ols=OLS(ytrain,xcon)
model=ols.fit()
model.summary()
```

```python
model.pvalues.sort_values().index[-1]
```

```python
adj=[]
Q=range(0,86,1)
```

```python
for i in Q:
    col_drop=model.pvalues.sort_values().index[-1]
    Xnew=Xnew.drop(labels=col_drop,axis=1)
    xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=21)
    from statsmodels.api import OLS,add_constant
    xcon = add_constant(xtrain)
    ols = OLS(ytrain,xcon)
    model = ols.fit()
    adj.append(model.rsquared_adj)
```

By plotting graph check the best rsquared_adj

# Linear Model

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(xtrain,ytrain)
pred_tr = model.predict(xtrain)
pred_ts = model.predict(xtest)
from sklearn.metrics import mean_squared_error
tr_err = mean_squared_error(ytrain,pred_tr)
ts_err = mean_squared_error(ytest,pred_ts)
```

```python
print(round(tr_err),round(ts_err))
```

# Ridge

```python
lambdas=[]
b= 42.5
for i in range(0,500,1):
    b=b+0.001
    lambdas.append(round(b,2))
```

```python
tr =[]
ts =[]
for i in lambdas:
    from sklearn.linear_model import Ridge
    rr = Ridge(alpha=i)
    model = rr.fit(xtrain,ytrain)
    pred_tr = model.predict(xtrain)
    pred_ts = model.predict(xtest)
    from sklearn.metrics import mean_squared_error
    tr_err = mean_squared_error(ytrain,pred_tr)
    ts_err = mean_squared_error(ytest,pred_ts)
    tr.append(tr_err)
    ts.append(ts_err)
```

```python
plt.plot(lambdas,tr,c="blue")
plt.plot(lambdas,ts,c="red")
```

# Use K-fold model:
to get best alpha value

```python
from sklearn.model_selection import GridSearchCV
tg = {"alpha":lambdas}
rr = Ridge()
cv = GridSearchCV(rr,tg,scoring="neg_mean_squared_error",cv=4)
cvmodel = cv.fit(Xnew,Y)
cvmodel.best_params_
```
